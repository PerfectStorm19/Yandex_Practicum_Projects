Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

Прочитаем все файлы, выведем случайные 5 строк, общую информацию о датасете (вместе с информацией об абсолютном и относительном количестве пропусков), основные статистики по столбцам и количество явных дубликатов.

Для удобства работы весь процесс реализуем через функции.

```python
def nans(df):
    '''функция выводит количество пропусков в каждом столбце и их долю от общего количества данных в столбце'''
    
    count_missing = df.isna().sum()
    percent_missing = round(df.isna().sum() * 100 / len(df), 2)
    
    missing_value_df = pd.DataFrame({'num_missing': count_missing,
                                     'percent_missing': percent_missing})
    
    return missing_value_df.reset_index().rename(columns={'index':'column'})
```

```python
def info_dataframe(data, details=False):
    '''функция преобразует df.info() в датафрейм'''
    
    dfInfo = data.columns.to_frame(name='column')
    dfInfo['dtype'] = data.dtypes
    dfInfo['non_null_count'] = data.notna().sum()  
    dfInfo.reset_index(drop=True, inplace=True)
    
    if details:
        rangeIndex = (dfInfo['non_null_count'].min(), dfInfo['non_null_count'].min())
        totalColumns = dfInfo['column'].count()
        dtypesCount = dfInfo['dtype'].value_counts()
        totalMemory = dfInfo.memory_usage().sum()
        return dfInfo, rangeIndex, totalColumns, dtypesCount, totalMemory
    else:
        return dfInfo
```

```python
def all_info(df):
    '''функция выводит случайные 5 строк, информацию по датафрейму, включая пропуски и их долю, статистики и кол-во дубликатов'''
    
    display(df.sample(5))
    print(f'Датафрейм содержит {df.shape[0]} записей и {df.shape[1]} колонки')
    all_info_df = info_dataframe(df).merge(nans(df), on='column')
    display(all_info_df)
    display(df.describe()) 
    print(f'Количество явных дубликатов в датафрейме {df.duplicated().sum()}')
    for column in df.columns:
        print(f'Уникальные значения колонки {column}:')
        print(df[column].unique())
```

То есть, в датасете содержатся данные по открытию договоров с 1 октября 2013 года до 1 февраля 2020 года.

Таким образом, более 84% клиентов все еще с нами, их контракты не завершены.

## 4. Моделирование.

### 4.1. Оценка корреляции признаков.

Исследуем корреляцию с помощью коэффициента корреляции `Phik` - в отличие от прочих видов корреляции (Пирсона, Спирмена), этот учитывает в том числе связи и между категориальными признаками. Кроме того, он способен фиксировать не только линейные, но и более сложные зависимости в данных.

```python
# создаем таблицу корреляций, передаем параметр interval_cols, чтобы указать на непрерывные признаки

phik_table = total_data[num_columns + cat_columns + ['services', 'total_days', 'target']].phik_matrix(interval_cols=['monthly_charges', 'total_charges', 'total_days'])

# отрисовываем график
plot_correlation_matrix(phik_table.values, 
                        x_labels=phik_table.columns, 
                        y_labels=phik_table.index, 
                        vmin=0, vmax=1, color_map="Blues", 
                        title=r'Correlation $\phi_K$', 
                        fontsize_factor=1.5, 
                        figsize=(20, 20))
```

**Выводы**

Сильнее всего с целевой переменной коррелируют признаки `total_charges` (0.30), `monthly_charges` (0.23), `online_backup` (0.23) и `partner` (0.23), `streaming_movies` (0.22), `device_protection` (0.22) и `payment_method` (0.21), однако, эти корреляции нельзя назвать сильными.

Самая сильная корреляция наблюдается с созданным признаком `total_days` (количество дней, прошедших от открытия до закрытия договора на оказание услуг), показатель 0.37 говорит о важности этого признака для моделирования. Однако, важно отметить, что этот признак также достаточно значимо коррелирует и с прочими фичами (например, с `total_charges`).

Меньше всего выражена связь целевой переменной и признаков `gender` (0.01), `services` (0.05) и `dependents` (0.05), в принципе они играют для целевой переменной такую незначительную роль, что их можно и совсем не брать при обучении моделей. 

Сильная корреляция наблюдается между `monthly_charges` и `total_charges` (0.71). Это понятно и обосновано, так как размер полной суммы, полученой от клиента за все время действия договора напрямую зависит от ежемесячных/ежегодных трат, которые он совершал это время. Тут можно столкнуться с проблемой мультиколлинеарности, которая очень важна при обучении линейных моделей. Учитывая также сильную корреляцию `total_charges` и `total_days`, можно попробоваться совсем отказаться от колонки `total_charges`.

Так как у вспомогательного признака `services` сильная корреляция с `multiple_lines` и `internet_service`, то его брать в обучение моделей не стоит.

У `internet_service` сильная корреляция с `multiple_lines`, то есть в принципе тоже можно оставить в дальнейшем только один признак.

Мультиколлинеарность существует и между колонками `streaming_tv` и `streaming_movies`, по всей видимости из-за того ,что клиенты часто подключают обе услуги сразу и значения этих колонок схожи. На целевой признак сильнее воздействует `streaming_movies`, значит можно оставить ее. Обе колонки имеют сильную взаимосвязь с `monthly_charges` и `total_charges`.

В целом, для обучения моделей лучше всего выбирать переменные, которые имеют высокую корреляцию с целевой переменной и минимальную корреляцию с другими переменными.

- `total_charges`, так как очень значительно коррелирует с `total_days`, хотя по сути имеет аналог в виде `monthly_charges`,

Как было замечено ранее, в наших данных есть дисбаланс по количеству клиентов каждого класса: около 85% пользователей остались с нами на момент выгрузки данных, и только 15% нас покинули.
    
Дисбаланс классов сильно влияет на такую метрику как `Accuracy`, так как моделям становится "выгоднее" склоняться к классу с большим количеством примеров. А вот ключевая метрика исследования `ROC-AUC` к дисбалансу классов не чувствительна, поскольку зависит не от самих ответов, а от меток классов объектов при упорядочивании по этим ответам.

Существует несколько способов для борьбы с дисбалансом классов:

1. Использование взвешенных моделей, которые учитывают степень дисбаланса весов разных классов.
2. Использование алгоритмов, которые умеют работать с несбалансированными классами, например, методов регуляризации.
3. Ресэмплинг датасета: увеличение числа образцов меньшего класса (oversampling) или уменьшение числа образцов большего класса (undersampling).

Особенно чувствительно к дисбалансу классов дерево решений. Логистическая регрессия и бустинги также могут быть чувствительны, но могут быть настроены на устранение этой проблемы.

Наконец, разделим обучающую и тестовую выборку (для тестирования лучшей модели) в соотношении 3 к 1. Укажем параметр `stratify`, чтобы учесть дисбаланс классов и выборки сохранили пропорцию.

### 4.3. Кодирование и масштабирование признаков, формирование пайплайнов, подготовка вспомогательных функций.

Применение линейных и "деревянных" моделей возвращает нас к проблеме кодирования категориальных признаков, которые могут быть учтены моделями, только пройдя через какой-либо энкодер.

Для линейных моделей хорошо подходит OneHotEncoder(). Для деревянных моделей гораздо лучше подходит другой тип кодировщика OrdinalEncoder().

Кроме того, численные признаки также нужно отмасштабировать, чтобы привести к единой шкале значений - это можно сделать с помощью StandardScaler(). 

Предобработку данных будем проводить с помощью трансформеров, в которых последовательно будут изменяться сначала категориальные колонки, зачем числовые. В зависимости от модели нужны трансформер упаковывать в пайплайн и его будем подавать на вход GridSearchCV для перебора гиперпараметров.

```python
# определяем трансформер для линейной модели

column_transformer_linear = ColumnTransformer([
    ('encode', OneHotEncoder(handle_unknown='ignore', drop='first', sparse=False), cat_features),
    ('scale', StandardScaler(), num_features)
    ], remainder='passthrough')
```

```python
column_transformer_linear.transformers
```

```python
# и трансформер для деревянных моделей

column_transformer_wood = ColumnTransformer([('encode', OrdinalEncoder(
        categories = [features[col].unique().tolist() for col in cat_features], 
        dtype=np.int16, 
        encoded_missing_value=-1), cat_features),
    ('scale', StandardScaler(), num_features)
    ])
```

```python
column_transformer_wood
```

Введем вспомогательные списки для занесения в них интересующих нас параметров, чтобы в дальнейшем можно было обращаться в ним напрямую по индексам в случае необходимости.

```python
names = []
roc_auc_scores = []
accuracy_scores = []
fit_times = []
predict_times = []
```

Обозначим функцию, которая будет проводить операцию кросс-валидации по метрикам `accuracy` и `ROC-AUC` внутри поиска лучших гиперпараметров по сетке, а затем получать метрики с лучшей модели и оформлять их в датафрейм.

```python
def get_metrics_gridsearch(model, column_transformer, params, name):  
    '''функция возвращает метрики после поиска лучшей модели по сетке гиперпараметов и добавляет их в список'''
    
    df_metrics = pd.DataFrame()
    
    pipe = Pipeline([('preprocess', column_transformer),
                      ('model', model)])
    
    search = GridSearchCV(estimator=pipe,
                          param_grid=params,
                          scoring=['roc_auc', 'accuracy'],
                          refit='roc_auc',
                          cv=5,
                          n_jobs=-1)
    
    search.fit(features_train, target_train)
    
    results = search.cv_results_
    best_index = search.best_index_
    best_model = search.best_estimator_
    
    # значение метрик
    model_roc_auc = round(search.best_score_, 4)
    model_accuracy = round(search.cv_results_['mean_test_accuracy'][search.best_index_], 4)
                      
    # время обучения модели
    fit_time = results['mean_fit_time'][best_index]
    
    # время предсказания модели
    pred_time = results['mean_score_time'][best_index]
    
    df_metrics['Модель'] = [name]
    df_metrics['ROC_AUC'] = model_roc_auc
    df_metrics['Accuracy'] = model_accuracy                       
    df_metrics['Время обучения, сек'] = round(fit_time, 4)
    df_metrics['Время предсказания, сек'] = round(pred_time, 4)
              
    names.append(name)
    roc_auc_scores.append(model_roc_auc)
    accuracy_scores.append(model_accuracy)
    fit_times.append(fit_time)
    predict_times.append(pred_time)
    
    print('Параметры лучшей модели', search.best_params_)
    
    return best_model, df_metrics
```

Также нам потребуется функция визуализации важности признаков.

```python
def get_feature_importances(model, features):
    '''функция визуализирует важность признаков для модели'''
    
    importances = model.feature_importances_
    feature_importances = pd.Series(importances, index=features.columns)
    sorted_importances = feature_importances.sort_values()
    sorted_importances.plot(kind='barh', figsize=(15, 10))
    plt.xlabel('Название признака\n', fontsize=14, weight='bold')
    plt.ylabel('Важность', fontsize=14, weight='bold')
    plt.title('Важность признаков для модели\n', fontsize=14, weight='bold')
    plt.show()
```

Описанные кодировки были помещены в пайплайны с помощью `ColumnTransformer`, чтобы стандартизировать процесс преобразований, сделать его автономным и последовательным.

Результаты проведенных экспериментов и полученные метрики отражены в таблице, а также визуализированы на графиках `ROC-AUC` и сравнения по точности.
