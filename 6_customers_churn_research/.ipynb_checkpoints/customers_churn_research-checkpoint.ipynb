```python
# выведем количество пропусков в каждом столбце и их долю от общего количества данных в столбце

def nans(df):
    
    count_missing = df.isna().sum()
    percent_missing = round(df.isna().sum() * 100 / len(df), 2)
    
    missing_value_df = pd.DataFrame({'num_missing': count_missing,
                                     'percent_missing': percent_missing})
    
    return missing_value_df

nans(churn)
```

### Заполнение пропусков

```python
# выведем строки с NaN в интересующем нас столбце

churn.query('Tenure.isna()')
```

```python
# построим гистограмму, чтобы визуально оценить распределение признака

plt.title('Распределение значений признака "Tenure" до заполнения пропусков\n', fontsize = 22)
churn['Tenure'].hist();
```

```python
# попробуем найти корреляции признака с другими

plt.title('Корреляция данных в столбце "Tenure" относительно остальных столбцов\n', fontsize = 22)
sns.heatmap(churn.corr()[['Tenure']]
                 .sort_values(by='Tenure'), annot=True, fmt='.2f', cmap='jet');
```

Как видим, некой закономерности в расположении пропусков в датасете и взаимозависимостей с другими признаками не обнаружено. Вероятно, потеря данных могла быть спровоцирована техническим сбоем или некорректным переносом данных из разных источников. 


Так как таких пропусков в столбце `Tenure` у нас менее 10%, то их можно было бы просто удалить. Однако, остальные 13 столбцов с признаками клиентов заполнены без видимых багов, и было бы жаль полностью терять эти записи из-за одного столбца. 

Чтобы не удалять строки, но и не внести ложные зависимости в данные, лучше всего такие пропуски заполнить случайными значениями, сэмплированными из распределения значений признака `Tenure`.

(ВНИМАНИЕ: код далее не списан у коллег, а позаимствован у преподавателя Тимура Усатого из открытого обсуждения в Пачке)

```python
# для отбора записей с пропусками в Tenure
mask = churn["Tenure"].isna()

# сколько пропущенных значений всего нужно заполнить
n_samples = mask.sum()

# сэмплируем случайные значения, используем values, чтобы корректно записать значения в пропущенные ячейки (игнорируя индекс)
random_samples = churn.loc[~mask, "Tenure"].sample(n_samples, replace=True, random_state=RANDOM).values

# заполняем случайными значениями пропуски
churn.loc[mask, "Tenure"] = random_samples
```

Проверим правильность заполнения пропусков.

```python
churn.info()
```

```python
plt.title('Распределение значений признака "Tenure" после заполнения пропусков\n', fontsize = 22)
churn['Tenure'].hist();
```

Таким образом, нам удалось равномерно заполнить пропуски в столбце `Tenure`, не исказив распределение признака, а следовательно, и получаемые нами в процессе исследования выводы.

Теперь нам нужно подробнее рассмотреть содержимое колонок и выделить те фичи, на которых будет проходить обучение наших моделей, а также разделить данные на обучающую и тестовую выборки.

### Изучение и выделение признаков

- `RowNumber` — индекс строки в данных
- `CustomerId` — уникальный идентификатор клиента
- `Surname` — фамилия.

Они не несут никакой практической пользы для исследования и более того, модели могут переобучиться под эти значения и выдавать необъективный score в реальных ситуациях. Поэтому их следует удалить.

```python
# удалим колонки, которые не предоставляют информации для обучения моделей

churn = churn.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

Остальные колонки

### Изучение целевого признака, разделение выборок

Теперь оценим распределение клиентов по целевому признаку - ушел клиент (1) или остался (0).

```python
target.value_counts().plot(kind='bar', 
                           title='\nРаспределение по целевому признаку\n', 
                           grid=True, figsize=(10, 5));
```

В силу специфики наших данных, наблюдается дисбаланс классов: оставшихся клиентов в 4 раза больше, чем тех, кто покинул банк. В жизни это хорошо, а в данных не очень, так как наши модели будут стремиться выдавать в качестве предсказаний значение из превалирующего класса для получения лучшего accuracy.

Чтобы дисбаланс не влиял на разбиение по обучающей и тестовой выборке, укажем в качестве параметра stratify=target, чтобы во всех выборках соотношение между классами было одинаковое.

Теперь поделим данные на обучающую, валидационную и тестовую выборки в соотношении 70:15:15. 

Так 70% изначального количества данных мы будем использовать для обучения различных моделей, 15% валидационных данных будут скрыты от моделей то стадии проверки моделей с лучшими гиперпараметрами, а еще 15% тестовых данных мы покажем только самой лучшей модели в самом конце. 

```python
# отделяем 70% обучающую выборку
features_train, features_control, target_train, target_control = train_test_split(features, 
                                                                        target, 
                                                                        test_size=0.3, 
                                                                        stratify=target, 
                                                                        random_state=RANDOM)

# делим валидацию и тестовые выборки пополам
features_val, features_test, target_val, target_test = train_test_split(features_control,
                                                                        target_control,
                                                                        test_size=0.5,
                                                                        stratify=target_control,
                                                                        random_state=RANDOM)

# выведем размеры полученных выборок
print(features_train.shape, features_val.shape, features_test.shape)
print(target_train.shape, target_val.shape, target_test.shape)
```

### Кодирование и масштабирование признаков

Теперь, когда выборки полностью разделены, и мы исключили утечку данных из обучающей выборки в тестовую (о которой модель во время обучения знать ничего НЕ должна), можно перейти к кодированию категориальных признаков и масштабированию численных. 

Библиотека sklearn предлагает широкие возможности по предобработке данных, таким образом, чтобы нам не нужно было вручную прописывать различные процессы.
В данном случае, мы воспользуемся двумя полезными функциями

`OneHotEncoder` https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html


`StandartScaler` https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
 
с помощью которых проведем все трансформации.

```python
# выделяем категориальные колонки для one-hot кодирования
categorical = features_train.select_dtypes(include='object').columns.to_list()
categorical
```

```python
# задаем энкодер для кодирования категориальных признаков

ohe = OneHotEncoder(drop='first', # определяет метод удаления категории фичи (удалит первую колонку из закодированных фичей)
                    handle_unknown='error', # определяет то, как мы поступим с неизвестными признаками во время трансформации
                    sparse=False) # устанавливаем в False, чтобы на выходе мы получили массив, а не матрицу

# обучаем энкодер на категориальных колонках тренировочной выборки
ohe.fit(features_train[categorical])

# добавляем закодированные признаки в features_train, метод get_feature_names_out() позволяет извлечь названия фичей для трансформации
features_train[ohe.get_feature_names_out()] = ohe.transform(features_train[categorical])

# удаляем изначальные колонки c категориальными признаками
features_train = features_train.drop(categorical, axis=1)
features_train
```

```python
# выделяем численные колонки для масштабирования
numeric = features_train.columns.tolist()[:-5]
numeric
```

```python
# задаем скейлер для стандартизации значений признаков
scaler = StandardScaler()

# можно обучить его на тренировочной выборке и сразу трансформировать ее с помощью fit_transform
features_train[numeric] = scaler.fit_transform(features_train[numeric])
features_train
```

Таким образом, все уникальные значения категориальных фичей разнесены по отдельным колонкам, а численные признаки приведены к единому масштабу. 

Подготовим валидационную и тестовую выборки аналогичным образом.

```python
features_val[ohe.get_feature_names_out()] = ohe.transform(features_val[categorical])
features_val = features_val.drop(categorical, axis=1)
features_val[numeric] = scaler.transform(features_val[numeric])

features_test[ohe.get_feature_names_out()] = ohe.transform(features_test[categorical])
features_test = features_test.drop(categorical, axis=1)
features_test[numeric] = scaler.transform(features_test[numeric])
```

## Исследование задачи

После проведения первичной предобработки данных вернемся к цели нашего исследования - спрогнозировать отток клиентов. 

Для задач классификации можно использовать 3 изученные нами ранее типа моделей:

- логистическую регрессию / logistic regression,
- дерево решений / decision tree,
- случайный лес / random forest (ансамбль деревьев).

Каждая из этих моделей обладает своими плюсами и минусами, мы будем рассматривать их последовательно, подбирать гиперпараметры с помощью 

`GridSearchCV` https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

и оценивать точность их предсказаний с помощью метрик accuracy (точность), recall (полнота) и F1-мера (аггрегирующая метрика).

Пайплайн исследования сделаем следующим:

- объявим объект модели соответствующего класса, 
- переберем гиперпараметры по сетке таким образом, чтобы модель училась и валидировала свои результаты через встроенную в метод кросс-валидацию,
- отберем модель с лучшими гиперпараметрами и финально провалидируем ее на отложенной валидационной выборке.

### Подготовка вспомогательных функций.

Так как пайплайн будет выглядеть однотипно для различных моделей, то есть смысл реализовать его через функции, чтобы не копировать большие куски кода, выполняющего одни и те же рутинные операции.

```python
def best_model_gridsearch(features_train, target_train, features_val,
                          model, parameters):
    '''Функция по нахождению лучшей модели по сетке параметров'''
    
    # поиск лучших гиперпараметров через кросс-валидацию с подаваемой моделью и перебираемыми параметрами
    search = GridSearchCV(model,
                          param_grid=parameters,
                          cv=5,
                          scoring='f1'
                         )
    
    # обучение на обучающей выборке
    search.fit(features_train, target_train)

    # вызываем параметры лучшей модели и ее лучший F1-score 
    print('Параметры лучшей модели:', search.best_estimator_)
    print('Значение F1-меры на лучшей модели', search.best_score_)
    
    # сохраняем и возвращаем лучшую модель
    best_model = search.best_estimator_
    return best_model
```

```python
def make_preds_show_metrics(best_model, features_val, target_val):
    '''Функция делает предсказания на лучшей модели и выводит все метрики на валидации'''
    
    # делаем предсказания модели с лучшими гиперпараметрами на валидации и получаем вероятности классов
    predictions_val = best_model.predict(features_val)
    probabilities_val = best_model.predict_proba(features_val)[:, 1]
    
    # получаем важные для нас метрики f1, roc-auc, а также FPR и TPR для построения ROC-AUC Curve
    f1 = f1_score(target_val, predictions_val)
    auc_roc = roc_auc_score(target_val, probabilities_val)
    fpr, tpr, thresholds = roc_curve(target_val, probabilities_val)
    
    # выводим полное описание метрик
    print (f'Accuracy = {accuracy_score(target_val, predictions_val)}',
           f'Precision = {precision_score(target_val, predictions_val)}',
           f'Recall = {recall_score(target_val, predictions_val)}',
           f'F1-score = {f1}',
           f'AUC-ROC = {auc_roc}',
           sep='\n')
    
    return f1, auc_roc, fpr, tpr, thresholds
```

```python
def roc_auc_curve(fpr_logreg, tpr_logreg,
                  fpr_dectree, tpr_dectree,
                  fpr_randfor, tpr_randfor):
    '''Функция для построения и сравнительного анализа ROC-AUC кривых'''
    
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    
    # все три кривых от разных моделей отрисовываются на одном графике
    plt.plot(fpr_logreg, tpr_logreg)
    plt.plot(fpr_dectree, tpr_dectree)
    plt.plot(fpr_randfor, tpr_randfor)
    
    plt.xlabel('\nFalse Positive Rate')
    plt.ylabel('True Positive Rate\n')
    plt.title('\nROC-кривые\n')
    plt.legend(('Baseline',
                'Логистическая регрессия',
                'Решающее дерево',
                'Случайный лес'),
                loc='upper left') 
    plt.show()
```

### Обучение моделей с предварительной процедурой Upsampling.

Преобразование проходит в несколько этапов:
    
- Разделить обучающую выборку на отрицательные и положительные объекты;
- Скопировать несколько раз положительные объекты;
- С учётом полученных данных создать новую обучающую выборку;
- Перемешать данные: идущие друг за другом одинаковые вопросы не помогут обучению.

```python
def upsample(features, target, repeat):
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=RANDOM)
    
    return features_upsampled, target_upsampled
```

Зададим параметр repeat с учетом того, что положительных объектов у нас в 4 раза меньше, чем отрицательных.

```python
features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
features_upsampled.shape, target_upsampled.shape
```

### Обучение моделей с предварительной процедурой Downsampling.

Преобразование проходит в несколько этапов:
    
- Разделить обучающую выборку на отрицательные и положительные объекты;
- Случайным образом отбросить часть из отрицательных объектов;
- С учётом полученных данных создать новую обучающую выборку;
- Перемешать данные, так как положительные не должны идти следом за отрицательными (алгоритмам будет сложнее обучаться).

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=RANDOM)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=RANDOM)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=RANDOM)
    
    return features_downsampled, target_downsampled
```

Исходя из пропорции дисбаланса классов, снизим количество объектов отрицательного класса в 4 раза, поставив параметр fraction=0.25.

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
features_downsampled.shape, target_downsampled.shape
```

```python
# проверим сбалансированность классов
target_downsampled.value_counts()
```
